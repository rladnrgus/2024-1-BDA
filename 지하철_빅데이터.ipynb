```python
subway = pd.read_csv('서울시 지하철 호선 별 역별 시간대별 승하차 인원정보.csv',encoding='EUC-KR')

# 승차 관련 컬럼만 필터링
in_subway= subway.columns.tolist()[:3]
for a in subway.columns.tolist():
    if a.find('승차')!= -1 :
        in_subway.append(a)

in_subway_df = subway[in_subway]

in_subway_df.head()
```

```python
station_group=in_subway_df.groupby(['호선명','지하철역']).sum()
station_group=station_group.drop(['사용월'],axis=1)
```
